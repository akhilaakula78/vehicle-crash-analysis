Selecting MinMaxScalar for Scaling:
For these data, MinMaxScaler was chosen as the scaling method. MinMaxScaler scales the features to a defined range (0-1) regardless of the original distribution.

This allows all features to contribute more equally to the regression process removing the impact of skewed features.
Since data distributions are not normal, MinMaxScaler is a better option as StandardScaler assumes normality. MinMaxScaler ensures equal treatment of all features by scaling the data to a specified range.
This scaler is more suitable as it compresses the data into a predetermined range regardless of the original distribution.
Using MinMaxScaler, we scale our dataset so that it is duly scaled to train machine learning models.
